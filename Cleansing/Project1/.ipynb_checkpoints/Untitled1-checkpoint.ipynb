```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
match_path= "Resources/Matches_Results.csv"
odds_path= "Resources/Matches_Odds.csv"
```

```python
matches=pd.read_csv(match_path)
odds=pd.read_csv(odds_path)
```

```python
matches.head(10)
```

```python
merged_df=pd.merge(matches, odds, on=["match_id"])
```

```python
merged_df=merged_df.drop(columns=["date_start_y"])
```

```python
merged_df
```

```python
#grouping the data for just the 5 leagues of interest
#also grouping datapoints of the same match id together

leagues = merged_df.groupby('competition_name_x')

English = leagues.get_group('Eng. Premier League')
German = leagues.get_group('German Bundesliga')
French = leagues.get_group('French Ligue 1')
Spanish = leagues.get_group('Spanish Liga Primera')
Italian = leagues.get_group('Italian Serie A')

eng_matches = English.groupby('match_id')
ger_matches = German.groupby('match_id')
fre_matches = French.groupby('match_id')
spa_matches = Spanish.groupby('match_id')
ita_matches = Italian.groupby('match_id')

eng_matches.head(1)

#print(f"{len(eng_matches)}, {len(ger_matches)}, {len(fre_matches)}, {len(spa_matches)}, {len(ita_matches)}")
#print(f"{len(English)}, {len(German)}, {len(French)}, {len(Spanish)}, {len(Italian)}")

#prints show that we are missing some matches since the leagues had 380 matches each, except the German one with 306
#of interest would be finding out which matches are missing from dataset
```
